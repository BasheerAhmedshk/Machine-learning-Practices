### Logistic Regression

Logistic Regression is a supervised machine learning algorithm used for classification tasks. It is a linear model that predicts the probability of a given class based on the features of the input data. 

Logistic Regression uses the sigmoid function to map the predicted probability to a range between 0 and 1.

Logistic Regression is a popular choice for binary classification problems, where the goal is to predict whether a given instance belongs to one class or another.

#### Types of Logistic Regression:

- Binary Logistic Regression: Logistic Regression is used for binary classification problems, where the goal is to predict whether a given instance belongs to one class or another.

- Multiclass Logistic Regression: Logistic Regression is used for multiclass classification problems, where the goal is to predict the class of a given instance.

- Logistic Regression with Multiple Classes: Logistic Regression is used for problems with multiple classes, where the goal is to predict the class of a given instance.

- ordinal Logistic Regression: Logistic Regression is used for ordinal classification problems, where the goal is to predict the rank or order of a given instance.


#### Uses of Logistic Regression:

- Email Spam Filtering: Logistic Regression can be used to classify emails as spam or not spam.

- Customer Churn Prediction: Logistic Regression can be used to predict whether a customer will churn or not.

- Fraud Detection: Logistic Regression can be used to detect fraudulent transactions.

- Sentiment Analysis: Logistic Regression can be used to predict the sentiment of a given text.

- Image Classification: Logistic Regression can be used to classify images into different categories.

- Medical Diagnosis: Logistic Regression can be used to predict the diagnosis of a medical condition.

- Credit Card Fraud Detection: Logistic Regression can be used to detect fraudulent credit card transactions.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('insurance.csv')
data.head(5)
```

```python
plt.scatter(data['age'], data['bought_insurance'], marker='x')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Features and target
X = data[['age']]
y = data['bought_insurance']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2)
```

```python
# Train model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
```

```python
model.score(X_test, y_test)
```

```python
model.predict_proba(X_test)
```

```python
import math 

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```
